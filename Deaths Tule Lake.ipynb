## Tule Lake Camp

Due to the tension between the United States and Japan (The Empire of Japan) during World War II, especially after the Pearl Harbor incident on December 7, 1941, on February 19, 1942, President Franklin D. Roosevelt signed the Executive order 9066 [3].

 This order authorized the Secretary of War to relocate all Japanese Americans residing on the West Coast to be interned in isolated camps, one of which is the Tule Lake Internment camp. Tule Lake camp was one of the 10 camps where Japanese Americans were incarcerated from 1942, to 1946. It was officially opened May 26, 1942 detaining Japanese Americans who lived on the West Coast from Washington, Oregon and North California.  In the camp were found hospitals, cemeteries, and farmland. Tule Lake was more like a prison-life atmosphere where Japanese Americans were forced to answer loyalty questionnaires. 

Due to the nature of these incidents at Tule Lake Camp, the focus of our project is to digitize all physical information available on people from this camp. By looking at death certificate records, we will be observing the death trends of each of these individuals who have passed away at this camp, along with determining the overall death tolls, age of each individual, their genders, death date, and birthplace.
